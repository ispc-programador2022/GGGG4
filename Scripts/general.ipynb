```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
etiquetas = ['Banco Nación', 'Dólar Libre', 'Dólar Mayorista', 'Dólar Bolsa', 'contado con Liqui']
compra = [161.00, 304.00, 162.27, 309.17, 322.63]
venta = [169.00, 308.00, 162.47, 309.34, 325.54]

co = np.arange(len(compra))
an = 0.35

fig, ax = plt.subplots()
ax.bar(co - an/2, compra, an, label='compra', color=['green'])
ax.bar(co + an/2, venta, an, label='venta', color=['red'])
ax.set_title('DólarHoy 16/11') #Titulo de la grafica
ax.set_ylabel('compra y venta') #Titulo eje y
ax.set_xticks(co)
ax.set_xticklabels(etiquetas, rotation= 60) #etiquetas/titulos en el eje x


plt.legend()
```

```python
from datetime import datetime
import matplotlib.pyplot as plt
import numpy as np

fecha = datetime.now().date()
dolarHoy = pd.read_csv(f"../data/diario/{fecha}-dolarHoy.csv")
mi_diccionary = dolarHoy.to_dict()

lista = []
for val in mi_diccionary.values():
    lista.append(val)
    
entidades = [x for x in lista[0].values()]
compra = [x for x in lista[1].values()]
venta = [x for x in lista[2].values()]

co = np.arange(len(compra))
an = 0.35

fig, ax = plt.subplots()
ax.bar(co - an/2, compra, an, label='compra')
ax.bar(co + an/2, venta, an, label='venta')
ax.set_title('DólarHoy 16/11')
ax.set_ylabel('compra y venta')
ax.set_xticks(co)
ax.set_xticklabels(entidades, rotation= 60)
plt.legend()
```
